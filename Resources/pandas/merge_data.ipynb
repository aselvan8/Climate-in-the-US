```python
import pandas as pd
import numpy as np
```

```python
path = "../../data/carbon_emission_by_sector.csv"
transp_df = pd.read_csv(path)
transp_df.dropna().head()
```

```python
new_df = transp_df.reset_index(drop=True).set_index("Sector")
```

```python
new_df1 = new_df.T
new_df2 = new_df1.reset_index()
new_df2.head()
```

```python
new_df2.rename(columns={"index": "year", "Total U.S. CO2 Emissions from energy use": "total_emission", "Transportation": "transportation_emission",
                       "Coal": "coal_emission", "Natural gas": "natural_gas_emission", "Electricity": "electricity_emission",
                       "Petroleum": "petroleum_emission", "Motor gasoline": "motor_emission", "Hydrocrabon gas liquidsa":"hydrocrabon_gas_emission",
                       "Jet fuel": "jet_fuel_emission", "Distillate fuel oil": "distillate_fuel_emission", "Residual fuel oil": "residual_fuel_emission",
                       "Lubricants": "lubricant_emission", "Aviation gas": "aviation_gas_emission", "Industrial": "industrial_emission", 
                        "Residential": "residential_emission", "Commercial": "commercial_emission"}, inplace=True)
```

```python
new_df2["year"] = new_df2["year"].astype(int)
new_df2.head()
```

```python
path1 = "../../data/renewable_energy_consumption.csv"
carbon_df = pd.read_csv(path1)
carbon_df1 = carbon_df.dropna()
carbon_df1.rename(columns={"Annual Total": "year"}, inplace=True)
carbon_df1.columns = carbon_df1.columns.str.lower()
carbon_df1.rename_axis("index").head()
carbon_df1.head()
```

```python
path2 = "../../data/worldbank_data.csv"
wb_df = pd.read_csv(path2)
```

```python
wb_df2 = wb_df.reset_index(drop=True).set_index("Series Name")
wb_df3 = wb_df2.T
wb_df4 = wb_df3.drop(["Country Name","Country Code","Series Code"])
wb_df4.columns
```

```python
wb_df4.rename(columns={"GDP (current US$)": "gdp_usd",
                       "GDP growth (annual %)": "gdp_growth_annual%", 
                       "GDP per unit of energy use (PPP $ per kg of oil equivalent)": "gdp_per_energyuse",
                       "Pump price for gasoline (US$ per gal)": "gasprice_per_gal",
                       "Inflated 2020 price for gasoline (US$ per gal)": "inflated_gaspr"
                       "CO2 emissions (metric tons per capita)": "emission_per_capita",
                       "Electric power consumption (kWh per capita)": "electric_consumption_per_capita",
                       "GDP per capita (current US$)": "gdp_per_capita_usd",
                       "GDP per capita growth (annual %)": "gdp_per_capita_annualgrowth%"
                      }, inplace=True)
wb_df5 = wb_df4.reset_index()
wb_df5.rename(columns={"index": "year"}, inplace=True)
wb_df5.rename_axis("index").head()
```

```python
wb_df5["year"] = wb_df5["year"].astype(int)
wb_df5.head()
```

```python
merge_df = pd.merge(left=new_df2, right=carbon_df1, left_on='year', right_on='year')
merge_df.head()
```

```python
merge_df2 = pd.merge(left=merge_df, right=wb_df5, left_on='year', right_on='year')

merge_df2.head()
```

```python
merge_df2.to_csv("../../data/cleaned/merged_energy_emission_data.csv")
```
